```python
import seaborn as sns
sns.get_dataset_names() 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
```

```python
# 1. scatter plot

sns.scatterplot(data=tips, x="total_bill", y="tip",hue="time", size="size", palette="deep")
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
# 2. line plot: Total bill by party size, split by sex

sns.lineplot(data=tips, x="size", y="total_bill", hue="sex", ci= None, marker="o")
plt.title("Line plot : Total Bill by Party Size and sex")
plt.show()
```

```python
# line plot

sns.lineplot(data=tips, x= 'size', y='total_bill', hue='sex',markers='o')
plt.title("Lineplot of Total Bill vs Size")
plt.show()
```

```python
tips.columns
```

```python
#3. bar plot : Avarage bill by day,split by sex

sns.barplot(data=tips, x="day", y="total_bill", hue = "sex",palette="muted")
plt.title("Bar plot : Avarage Total Bill by Day and Sex")
plt.show()
```

```python
# 4. Boxplot 

sns.boxplot(data=tips, x='day', y='tip', hue='smoker', palette='Set2')
plt.title("Box plot: Tips by Day and Smoker Status")
plt.show()
```

```python
# 5. violin plot: Total Bill by Day and Time

sns.violinplot(data=tips, x='day', y='total_bill', hue='time', split=True, palette='pastel')
plt.title("Violin Plot:Total Bill by Day and Time")  
plt.show()
```

```python
#6. count plot : Orders by Day, split by smoker status

sns.countplot(data=tips, x='day', hue='smoker', palette='dark')
plt.title("Count Plot: Orders by Day and Smoker Status")
plt.show()
```

```python
#7. regression plot : Total Bill vs Tip with regression line

sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title("Regression Plot of Total Bill vs Tip")   
plt.show()
```

```python
# 8. Histogram of total bill with KDE

sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='blue')
plt.title("Histogram: Distribution of Total Bill")
plt.show()
```

```python
#9. pair plot: Relationships between numeric variables

sns.pairplot(tips, hue='sex', vars=["total_bill", "tip", "size"], palette='husl')
plt.suptitle("pair plot: numerical variables by sex", y=1.02)
plt.show()
```

```python
# 10 catplot (point plot): Tips by Day and Sex

sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='point', palette='bright')
plt.title("catplot(point):Tips by day and Sex")
plt.show()
```

```python
# 11. joint plot: Total Bill vs Tip with marginal distributions

sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker', color='purple', palette='coolwarm')
plt.suptitle("Joint plot: Total Bill vs Tip By smoker", y=1.02)
plt.show()
```

```python
# FacetGrid : Total Bill by Day, facted by Time and Smoker

g = sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True)
g.map(sns.histplot, 'total_bill', bins=15)
g.fig.suptitle("FacetGrid: Total Bill by Time and Smoker", y=1.02)
plt.show()
```

```python
#13. strip plot: Tips by Day, color-coded by Sex
plt.figure(figsize=(8, 6))
sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True, palette='Set1')
plt.title("strip plot: Tips by data and gender")
plt.show()
```

```python
# 14. KDE PLOT: Total Bill density by Sex
plt.figure(figsize=(8, 6))
sns.kdeplot(data=tips, x='total_bill',hue='sex', fill=True, palette='tab10')
plt.title("kde plot:Total bill density by Sex")
plt.show()
```

# Explanation of plot
  
  1) Scatter plot: shows the relationship between total_bill and tip, with time and size as additional dimensions.

  2) Line plot: Trends of total_bill across size,split by sex.

  3) Bar plot: Average total_bill by day, split by sex.

  4) Box plot: Distribution of tip by day, split by smoker.

  5) violin plot: Distribution of total_bill by day,split by time.

  6) Count plot: Counts of orders by day, split by smoker.

  7) Regression plot: Linear fit between total_bill and tip.

  8) Histogram: Distribution of total_bill with a KDK curv

  9) pair plot: pairwise relationships between total_bill,tips and size, colored by sex.

  10) Catplot (point): point estimates of tip by day and sex.

  11) joint plot: total_bill vs tip with marginal histograms, colored by smoker.

  12) FaceGrid: total_bill histograms faceted by time and smoker.

  13) Strip plot: individual tip points by day, colored by sex.

  14) KDK plot: Density of total_bill split by sex.
