```python
import matplotlib.pyplot as plt
import seaborn as sns

sub = iris[iris['Species'] == 'Iris-setosa']

plt.figure(figsize=(8,6))

sns.kdeplot(
    data=sub,
    x='SepalLengthCm',
    y='SepalWidthCm',
    fill=True,
    cmap='plasma'
)

plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')

plt.show()
```
