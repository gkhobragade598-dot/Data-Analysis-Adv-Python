```python
nums = [12,15,18,21,26]

for num in nums:
    if num % 5 == 0:
        print(num)
```

```python
nums = [7,14,18,21,23,27] # hear there is no number which is divisible by 5 we got output as blank

for num in nums:
    if num % 5 == 0:
        print(num)
        break
else:

    print('Number Not Found') # hear else we dont write in if block but we can write in for block only
```
